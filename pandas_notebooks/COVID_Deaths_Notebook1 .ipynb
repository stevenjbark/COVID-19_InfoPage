```python
total_list = []

for index, row in covid_df1.iterrows():
    diff_list = [row[0], row[1], row[2], row[3]]
    for n in range(5,len(covid_df1.columns)):
        diff_list.append(row[n]-row[n-1])
    total_list.append(diff_list)
```

```python
#total_list[:10]
```

```python
len(total_list)
```

```python
covid_df2 = pd.DataFrame(total_list)
```

```python
covid_df2.head()
```

```python
covid_df2["County"] = ""
covid_df2.head()
```

```python
county_list = []
for index, row in covid_df2.iterrows():
    county_list.append(row[1].split("County")[0])
```

```python
county_list[:10]
```

```python
covid_df2["County"] = county_list
```

```python
covid_df2.head()
```

```python
# Removing the stateFIPS and the combined county names. Will use the extracted County names.
covid_df3 = covid_df2.drop([1], axis=1).drop([3], axis=1)
```

```python
# Remove the statewide Unallocated from County
covid_df4 = covid_df3.loc[covid_df3["County"] != "Statewide Unallocated"]
```

```python
# Setting County to be the index
covid_df5 = covid_df4.set_index("County")
```

```python
covid_df5.head()
```

```python
renamed_covid_df = covid_df5.rename(columns = {0: "countyFIPS", 2: "State"})
```

```python
renamed_covid_df.head()
```

```python
renamed_covid_df.to_csv("20200607_COVID_DiffMapDeaths.csv")
```
