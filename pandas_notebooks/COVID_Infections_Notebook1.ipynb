### Conversion of total confirmed infections data into normalized number of infections per 100,000. Strategy: data in confirmed infections is in numbers, divide by total population in county, then multiply by 100,000.

### Dataframe for confirmed cases.

```python
covid_dfA = pd.read_csv("20200607covid_confirmed_usafacts.csv")
```

```python
covid_dfA.head()
```

```python
covid_dfA["population"] = ""
```

### Dataframe for county populations

```python
countypop_df = pd.read_csv("20200607covid_county_population_usafacts.csv")
```

```python
countypop_df.head(5)
```

### Add populations from countypop_df into covid_dfA.

```python
covid_dfA["population"] = countypop_df["population"]
```

### Remove Statewide Unallocated to eliminate population = 0 errors. 

```python
covid_dfB = covid_dfA.loc[covid_dfA["County Name"] != "Statewide Unallocated"]
```

```python
covid_dfB.loc[covid_dfB["population"] == 0]
```

### Remove the rows with 0 for population. Census and unallocated are not populations, cruise ship is not a population. 

```python
covid_dfC = covid_dfB.loc[covid_dfB["population"] != 0]
```

```python
covid_dfC.shape
```

### New dataframe with calculation for infection number data for each day with population normalization per 100,000.

```python
columns = covid_dfC.columns
```

```python
len(columns)
```

```python
normalized_list = []

for index, row in covid_dfC.iterrows():
    new_list = [row[0],row[1], row[2], row[3]]
    for n in range(4, len(columns)):
        x = row[n]/row[len(columns)-1]
        new_list.append(100000*x)
    normalized_list.append(new_list)
    
```

```python
# Check that the lists within the normalized list added up to 3142 as in shape for rows in covid_dfC
len(normalized_list)
```

```python
# Checked that the numbers were reasonable for list of lists!
# normalized_list[96]
```

### Create dataframe from the list of lists.

```python
covid_dfD = pd.DataFrame(normalized_list, columns=columns)
```

```python
covid_dfD.head()
```

### Write to a .csv file as normalized per 100000 data.

```python
covid_dfD.to_csv("20200607_NormalizedInfectionData.csv")
```

# Create new dataframe and file for differences in infections each day of outbreak.

```python
covid_df1 = pd.read_csv("20200607covid_confirmed_usafacts.csv")
```

```python
covid_df1.shape
```

```python
covid_df1.head()
```

```python
len(covid_df1.columns)
```

### We are calculating differences, so all of the columns will be present except the first day (1/22/20) because there are no days before that one. Remove.

```python
columns = list(covid_df1.columns)
columns.pop(4)
```

### Create empty list, then append new lists from countyFIPS, County Name, State, stateFIPS, then add new item for number of infections on each day minus the number of infections from previous day. Then divide by previous day to indicate percentage increase or decrease. If previous day was 0, then input 0 instead of calculating (division by 0).

```python
total_list = []

for index, row in covid_df1.iterrows():
    diff_list = [row[0], row[1], row[2], row[3]]
    for n in range(5,len(covid_df1.columns)):
        if row[n] == 0:
            diff_list.append(0)
        else:
            diff_list.append(((row[n]-row[n-1])/row[n])*100)
    total_list.append(diff_list)
```

```python
#total_list[:10]
```

```python
len(total_list)
```

```python
covid_df2 = pd.DataFrame(total_list, columns=columns)
```

```python
covid_df2.head()
```

### Remove Statewide Unallocated information, not useful.

```python
covid_df3 = covid_df2.loc[covid_df2["County Name"] != "Statewide Unallocated"]
```
